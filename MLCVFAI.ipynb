```python
from fastai import *
from fastai.vision import *
```

```python
import os
print(os.getcwd())
```

```python
#Define the classes
classes = ['looking_at_watch', 'looking_out','neutral','talking_over_phone', 'texting', 'touching_face']
```

```python
#set the path
path = Path('data_reduced')
print(path)
```

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.15,
         ds_tfms=get_transforms(), size=64, num_workers=0).normalize(imagenet_stats)
```

```python
#check all the classes set
data.classes
```

```python
data.show_batch(rows=5, figsize=(3,5))
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(5)
```

```python
learn.save('resnet_34_multiclass')
```

```python
learn.unfreeze()
```

```python
#doc(unfreeze())
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, max_lr=slice(1e-6,1e-5))
```

```python
learn.save('resnet34_6classes_dataReduced(1)')
```

```python
learn.fit_one_cycle(15, max_lr=slice(1e-6,1e-4))
```

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
interp.plot_top_losses(9, figsize=(15,11))
```

```python
doc(interp.plot_top_losses)
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```python
learn.export('resnet_34_multiClassActivity.pkl')
```

```python
path = os.getcwd()
fileList = os.listdir(path+"\\test")
for eachFile in fileList:
    img = open_image(path+"\\test\\"+str(eachFile))
    print('{0}==> predicts==> {1}'.format(eachFile,learn.predict(img)))
    print('\n')
```

```python
from fastai import *
from fastai.vision import *
```

```python
import os
print(os.getcwd())
```

```python
classes = ['LookingAtWatch', 'Neutral','TalkingOverPhone', 'Texting']
```

```python
#set the path
path = Path('database')
print(path)
```

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
         ds_tfms=get_transforms(do_flip=True), size=(224,224), num_workers=0).normalize(imagenet_stats)
```

```python
#check all the classes set
data.classes
```

```python
data.show_batch(rows=6, figsize=(15,12))
```

```python
resnet50_learner = cnn_learner(data, models.resnet50, metrics=error_rate)
```

```python
resnet50_learner.fit_one_cycle(4)
```

```python
resnet50_learner.save('resnet50_4C_itr5')
```

```python
resnet50_learner.unfreeze()
```

```python
del resnet50_learner
```

```python
torch.cuda.empty_cache()
```

```python
resnet50_learner.lr_find()
```

```python
resnet50_learner.recorder.plot()
```

```python
resnet50_learner.fit_one_cycle(4, max_lr=slice(1e-4,1e-2))
```

```python
resnet50_learner.export('resnet50_MCA_itr1.pkl')
```

```python
path = os.getcwd()
fileList = os.listdir(path+"\\testSamples")
for eachFile in fileList:
    img = open_image(path+"\\testSamples\\"+str(eachFile))
    print('{0}==> predicts==> {1}'.format(eachFile,resnet50_learner.predict(img)))
    print('\n')
```
